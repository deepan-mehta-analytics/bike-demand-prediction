# API Integration for Weather Data

In real-world data science applications, models are rarely used with static datasets. Instead, they rely on real-time or near real-time data sources.

In this project, weather data plays a crucial role in predicting bike-sharing demand. Therefore, we integrate an external API to fetch weather forecast data dynamically.

This notebook focuses on:

- Connecting to the OpenWeather API
- Sending HTTP requests to retrieve forecast data
- Parsing JSON responses into structured format
- Extracting relevant weather features

This step forms the foundation of the prediction pipeline, enabling the model to operate on live data.

## Import Required Libraries

We import libraries necessary for:

- Sending HTTP requests to APIs
- Parsing JSON responses
- Structuring data into tabular format
- Handling datetime values

```r
# Import requests library to send HTTP requests to API endpoints
import requests

# Import pandas for organizing data into DataFrame format
import pandas as pd

# Import json module to work with JSON responses
import json

# Import datetime for handling date and time values
from datetime import datetime
```

## Define API Parameters

To retrieve weather forecast data, we must define:

- API endpoint (URL)
- API key for authentication
- Query parameters such as city name and units

These parameters are used to construct the API request.

```r
# Define OpenWeather API endpoint for 5-day forecast
BASE_URL = "https://api.openweathermap.org/data/2.5/forecast"

# Define your API key (replace with your actual key)
API_KEY = "YOUR_API_KEY_HERE"

# Define city for which weather data will be fetched
CITY = "Seoul"

# Define query parameters for API request
params = {
    "q": CITY,             # City name
    "appid": API_KEY,      # API key for authentication
    "units": "metric"      # Use metric system (Celsius)
}
```

## Send API Request

We send a GET request to the API endpoint using the defined parameters.

The API returns weather forecast data in JSON format, which must be parsed for further processing.

```r
# Send HTTP GET request to API
response = requests.get(BASE_URL, params=params)

# Convert response into JSON format
weather_json = response.json()

# Print keys to understand structure of JSON response
weather_json.keys()
```

## Inspect JSON Structure

The API response contains nested JSON data. The key component is:

- `list`: contains forecast entries at different timestamps

Each entry includes:

- Temperature
- Humidity
- Wind speed
- Weather condition
- Timestamp

We extract these fields for further processing.

```r
# Extract forecast list from JSON response
forecast_list = weather_json.get("list", [])

# Display first record to inspect structure
forecast_list[0]
```

## Extract Relevant Weather Features

We iterate through each forecast record and extract relevant variables:

- Temperature
- Humidity
- Wind speed
- Weather condition
- Forecast datetime

These values are stored in a structured format for analysis.

```r
# Initialize empty list to store extracted weather records
weather_data = []

# Loop through each forecast entry
for entry in forecast_list:
    
    # Extract timestamp
    forecast_time = entry["dt_txt"]
    
    # Extract main weather metrics
    temperature = entry["main"]["temp"]
    humidity = entry["main"]["humidity"]
    
    # Extract wind speed
    wind_speed = entry["wind"]["speed"]
    
    # Extract weather condition (e.g., Clear, Rain)
    weather_condition = entry["weather"][0]["main"]
    
    # Append extracted data as dictionary
    weather_data.append({
        "datetime": forecast_time,
        "temperature": temperature,
        "humidity": humidity,
        "wind_speed": wind_speed,
        "weather": weather_condition
    })

# Convert list into DataFrame
weather_df = pd.DataFrame(weather_data)

# Display first few rows
weather_df.head()
```

## Convert Datetime Format

The datetime field is currently in string format. We convert it into a proper datetime object for further processing.

This allows extraction of:

- Hour
- Day
- Temporal features for modeling

```r
# Convert datetime column to pandas datetime format
weather_df["datetime"] = pd.to_datetime(weather_df["datetime"])

# Extract hour from datetime
weather_df["hour"] = weather_df["datetime"].dt.hour

# Display updated DataFrame
weather_df.head()
```

## Summary

In this notebook, we successfully:

- Connected to an external weather API
- Retrieved real-time forecast data
- Parsed nested JSON structures
- Extracted relevant features into a structured DataFrame
- Converted datetime into usable format
