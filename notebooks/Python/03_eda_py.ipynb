# Exploratory Data Analysis (EDA)

After transforming the dataset into a clean and structured format, the next step is to explore the data to uncover patterns, relationships, and trends.

Exploratory Data Analysis (EDA) is a critical phase in the data science workflow, as it helps build intuition about the dataset and informs feature selection for modeling.

In this notebook, we analyze how environmental and temporal variables influence bike-sharing demand using:

- Statistical summaries
- SQL-based aggregation queries
- Data visualization techniques

Key questions explored include:

- How does temperature affect bike demand?
- What is the relationship between humidity and usage?
- How does demand vary across hours and seasons?
- Are there observable trends that can guide model development?

## Import Required Libraries

We use the following libraries for analysis and visualization:

- `pandas`: data manipulation
- `matplotlib` and `seaborn`: visualization
- `sqlite3`: SQL-based querying on datasets

```r
# Import required libraries

import pandas as pd              # Data manipulation
import numpy as np               # Numerical operations
import matplotlib.pyplot as plt  # Plotting
import seaborn as sns            # Advanced visualization
import sqlite3                   # SQL queries on datasets

# Set visualization style
sns.set(style="whitegrid")
```

## Load Processed Dataset

We load the cleaned dataset produced in the previous ETL stage. This dataset contains structured numerical and categorical variables ready for analysis.

```r
# Load dataset

df = pd.read_csv("cleaned_bike_sharing.csv")

# Preview dataset
df.head()
```

## Basic Statistical Summary

Understanding summary statistics helps identify the distribution, central tendency, and variability of key variables.

```r
# Summary statistics

df.describe()
```

## SQL-Based Exploratory Analysis

SQL queries provide an alternative way to explore the dataset by performing aggregations and filtering operations.

This approach is particularly useful when working with large datasets or databases.

```r
# Create in-memory SQLite database

conn = sqlite3.connect(":memory:")

# Load dataframe into SQL table
df.to_sql("bike_data", conn, index=False, if_exists='replace')

# Query: Average bike demand by hour

query = """
SELECT hour, AVG(rented_bike_count) AS avg_demand
FROM bike_data
GROUP BY hour
ORDER BY hour
"""

hourly_demand = pd.read_sql(query, conn)

hourly_demand.head()
```

## Hourly Demand Trend

We visualize how bike demand varies throughout the day.

```r
# Plot hourly demand trend

plt.figure(figsize=(10,5))
plt.plot(hourly_demand['hour'], hourly_demand['avg_demand'])
plt.xlabel("Hour")
plt.ylabel("Average Bike Demand")
plt.title("Hourly Bike Demand Trend")
plt.show()
```

## Temperature vs Bike Demand

Temperature is one of the most influential factors affecting bike usage. We explore its relationship with demand.

```r
# Scatter plot: Temperature vs Demand

plt.figure(figsize=(8,5))
sns.scatterplot(x=df['temperature'], y=df['rented_bike_count'])
plt.title("Temperature vs Bike Demand")
plt.show()
```

## Humidity vs Bike Demand

Humidity can negatively impact outdoor activities. We analyze its relationship with bike usage.

```r
# Scatter plot: Humidity vs Demand

plt.figure(figsize=(8,5))
sns.scatterplot(x=df['humidity'], y=df['rented_bike_count'])
plt.title("Humidity vs Bike Demand")
plt.show()
```

## Correlation Analysis

Correlation analysis helps quantify the strength of relationships between variables.

```r
# Compute correlation matrix

corr_matrix = df.corr(numeric_only=True)

# Heatmap visualization
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

## Key Insights

From the exploratory analysis, several important patterns emerge:

- Temperature shows a strong positive relationship with bike demand
- Humidity tends to have a negative relationship with usage
- Demand varies significantly across different hours of the day
- Temporal patterns indicate peak usage during commuting hours

These insights will guide feature selection and model design in the next stage.
