# Data Wrangling and Transformation (ETL)

After collecting data from multiple sources, the next step is to clean and transform the data into a structured format suitable for analysis and modeling.

Raw data often contains:

- Missing values
- Inconsistent formatting
- Unnecessary characters or symbols
- Mixed data types

Data wrangling (ETL: Extract, Transform, Load) ensures that the dataset is:

- Clean
- Consistent
- Ready for analysis and modeling

This notebook focuses on transforming the collected bike-sharing and weather datasets using R.

## Import Required Libraries

We use the following libraries:

- `dplyr`: data manipulation
- `stringr`: string processing and regex operations
- `tidyr`: reshaping data
- `readr`: reading CSV files

```r
# Load required libraries

# dplyr is used for filtering, selecting, mutating, and grouping data
library(dplyr)

# stringr is used for string manipulation and regex operations
library(stringr)

# tidyr is used for reshaping and cleaning data
library(tidyr)

# readr is used for reading CSV files efficiently
library(readr)
```

## Load Raw Dataset

We load the dataset collected in the previous step.

This dataset may contain inconsistencies that need to be addressed before analysis.

```r
# Load dataset, verify the path to the csv file and edit the code below accordingly.
bike_data <- read_csv("SeoulBikeData.csv")

# View first few rows
head(bike_data)
```

## Standardize Column Names

Column names may contain spaces, special characters, or inconsistent formatting.

We standardize column names to:

- Lowercase
- Underscore-separated format
- Remove special characters

```r
# Convert column names to lowercase
colnames(bike_data) <- tolower(colnames(bike_data))

# Replace spaces with underscores
colnames(bike_data) <- str_replace_all(colnames(bike_data), " ", "_")

# Display updated column names
colnames(bike_data)
```

## Clean Data Using Regular Expressions

Some columns may contain unwanted text or symbols.

We use regular expressions (regex) to:

- Remove unwanted characters
- Extract numeric values
- Clean string-based columns

```r
# Example: Remove non-numeric characters from a column (if needed)

# Replace any character that is not a digit or decimal point
# This ensures numeric consistency

bike_data$temperature <- as.numeric(
  str_replace_all(bike_data$temperature, "[^0-9.]", "")
)
```

## Handle Missing Values

Missing values can negatively impact model performance.

We:

- Identify missing values
- Remove or impute them

```r
# Check missing values per column
colSums(is.na(bike_data))

# Remove rows with missing values
bike_data_clean <- na.omit(bike_data)

# Verify cleaned dataset
colSums(is.na(bike_data_clean))
```

## Convert Data Types

Some columns may not be in the correct format.

We ensure:

- Numerical variables are numeric
- Categorical variables are factors

```r
# Convert rented bike count to numeric
bike_data_clean$rented_bike_count <- as.numeric(bike_data_clean$rented_bike_count)

# Convert season to factor
bike_data_clean$seasons <- as.factor(bike_data_clean$seasons)

# Convert holiday to factor
bike_data_clean$holiday <- as.factor(bike_data_clean$holiday)
```

## Create Dummy Variables

Categorical variables must be converted into numerical form for modeling.

We create dummy variables using one-hot encoding.

```r
# Create dummy variables for categorical features

bike_data_dummies <- bike_data_clean %>%
  mutate(
    season_spring = ifelse(seasons == "Spring", 1, 0),
    season_summer = ifelse(seasons == "Summer", 1, 0),
    season_autumn = ifelse(seasons == "Autumn", 1, 0),
    season_winter = ifelse(seasons == "Winter", 1, 0)
  )
```

## Normalize Numerical Variables

Normalization ensures that all features are on a similar scale, which improves model performance.

We apply min-max scaling.

```r
# Define normalization function
normalize <- function(x) {
  (x - min(x)) / (max(x) - min(x))
}

# Apply normalization to selected columns
bike_data_normalized <- bike_data_dummies %>%
  mutate(
    temperature = normalize(temperature),
    humidity = normalize(humidity),
    wind_speed = normalize(wind_speed)
  )
```

## Final Clean Dataset

The dataset is now:

- Cleaned
- Standardized
- Ready for analysis and modeling

```r
# Display cleaned dataset
head(bike_data_normalized)

# Save cleaned dataset
write_csv(bike_data_normalized, "clean_bike_data.csv")
```

## Summary

In this notebook, we:

- Cleaned raw data
- Standardized column names
- Used regex for data cleaning
- Handled missing values
- Created dummy variables
- Normalized numerical features

These steps ensure that the dataset is ready for exploratory analysis and modeling.
