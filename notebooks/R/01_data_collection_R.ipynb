# Data Collection for Bike-Sharing Demand Prediction

In real-world data science projects, data is rarely obtained from a single source. Instead, multiple data sources must be integrated to build a comprehensive dataset.

In this project, bike-sharing demand prediction relies on combining:

- Weather data retrieved via APIs
- Global bike-sharing system information obtained through web scraping
- Structured datasets downloaded from cloud storage

As illustrated in the project overview, data is collected through:

- HTTP requests (API calls)
- Web scraping from HTML pages
- File downloads in CSV format

This notebook focuses on implementing these data collection techniques using R.

## Import Required Libraries

We use the following R libraries:

- `httr`: to make HTTP requests to APIs
- `jsonlite`: to parse JSON responses
- `rvest`: for web scraping HTML content
- `readr`: for reading CSV files
- `dplyr`: for data manipulation

```r
# Load required libraries

# httr is used to send HTTP requests to APIs
library(httr)

# jsonlite is used to parse JSON responses into R data structures
library(jsonlite)

# rvest is used for web scraping HTML pages
library(rvest)

# readr is used for reading CSV files efficiently
library(readr)

# dplyr is used for data manipulation
library(dplyr)
```

## Part 1: Web Scraping Bike-Sharing Systems (Wikipedia)

The global bike-sharing systems dataset is obtained from a Wikipedia page.

This dataset contains:

- Names of bike-sharing systems
- Cities and countries
- System details

We scrape the HTML table and convert it into a structured data frame.

```r
# Define Wikipedia URL containing bike-sharing systems
wiki_url <- "https://en.wikipedia.org/wiki/List_of_bicycle-sharing_systems"

# Read HTML content from the webpage
wiki_page <- read_html(wiki_url)

# Extract all tables from the page
tables <- wiki_page %>% html_table(fill = TRUE)

# Select the first table (main bike-sharing systems table)
bike_systems <- tables[[1]]

# Display first few rows
head(bike_systems)
```

## Part 2: API Data Collection (OpenWeather)

Weather data is collected using the OpenWeather API.

This data includes:

- Temperature
- Humidity
- Wind speed
- Forecast timestamps

We use HTTP requests to retrieve JSON data and convert it into a structured format.

```r
# Define API key (replace with your actual API key)
api_key <- "YOUR_API_KEY"

# Define city name
city <- "Seoul"

# Construct API URL for current weather
api_url <- paste0(
  "https://api.openweathermap.org/data/2.5/weather?q=",
  city,
  "&appid=", api_key,
  "&units=metric"
)

# Send GET request to API
response <- GET(api_url)

# Convert response content to text
content_text <- content(response, "text", encoding = "UTF-8")

# Parse JSON into R list
weather_data <- fromJSON(content_text)

# Inspect structure
str(weather_data)
```

## Extract Relevant Weather Features

We extract key variables from the API response for further analysis.

```r
# Extract temperature
temperature <- weather_data$main$temp

# Extract humidity
humidity <- weather_data$main$humidity

# Extract wind speed
wind_speed <- weather_data$wind$speed

# Combine into a data frame
weather_df <- data.frame(
  temperature = temperature,
  humidity = humidity,
  wind_speed = wind_speed
)

# Display extracted data
weather_df
```

## Part 3: Load Bike-Sharing Dataset (CSV)

The historical bike-sharing dataset is downloaded from cloud storage.

This dataset includes:

- Bike rental counts
- Weather variables
- Time-based variables

We load and inspect the dataset.

```r
# Load dataset from CSV file
bike_data <- read_csv("seoul_bike_sharing.csv")

# Display first few rows
head(bike_data)
```

## Inspect Dataset Structure

```r
# Display structure of dataset
str(bike_data)

# Summary statistics
summary(bike_data)

# Check missing values
colSums(is.na(bike_data))
```
